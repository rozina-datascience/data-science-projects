### Null & Alternate Hypothesis

**Question 1:**
Suppose you are conducting a study to determine whether a new drug is effective in reducing blood pressure. Define the null hypothesis (H0) and the alternate hypothesis (H1) for this study. Explain the significance of each hypothesis in the context of your research.

**Question 2:**
You are working with a manufacturing company that claims its new manufacturing process can produce car batteries with a longer lifespan than the current process. Formulate the null and alternate hypotheses for this scenario. Additionally, explain why it's essential to have clear and well-defined hypotheses before conducting hypothesis testing.

**Question 3:**
A fast-food restaurant chain is concerned that the average delivery time for online orders has increased in recent months, potentially affecting customer satisfaction. Create the null and alternate hypotheses for testing whether there has been a significant increase in delivery times.

*These questions will help you to practice defining and understanding null and alternate hypotheses in various real-world scenarios.*

```python
# Example: Hypothesis testing for blood pressure reduction after taking the new drug
import numpy as np
import scipy.stats as stats

# Example data: blood pressure reduction for 20 patients after taking the drug
bp_reduction = np.array([10, 8, 12, 9, 7, 15, 5, 14, 9, 10, 8, 11, 6, 7, 8, 9, 12, 5, 10, 8])

# Null hypothesis: mean reduction = 0
# Perform one-sample t-test
t_stat, p_val = stats.ttest_1samp(bp_reduction, 0)

# Print results
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_val}")

# Conclusion
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: The new drug is effective in reducing blood pressure.")
else:
    print("Fail to reject the null hypothesis: No significant evidence to conclude that the drug is effective.")
```

```python
# Example: Hypothesis testing for the lifespan of car batteries (two-sample t-test)
# Lifespan data for batteries from current and new manufacturing process
current_process = np.array([500, 550, 520, 530, 510, 560, 505])
new_process = np.array([530, 590, 570, 600, 610, 590, 595])

# Null hypothesis: mean(new_process) <= mean(current_process)
# Perform independent two-sample t-test (one-sided test)
t_stat, p_val = stats.ttest_ind(new_process, current_process)

# Print results
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_val}")

# Conclusion
alpha = 0.05
if p_val/2 < alpha and t_stat > 0:  # One-tailed test
    print("Reject the null hypothesis: The new process produces batteries with a longer lifespan.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the new process produces batteries with a longer lifespan.")
```

```python
# Example: Hypothesis testing for fast-food delivery times
# Delivery time data for two periods: before and after
before = np.array([30, 28, 32, 35, 33, 31, 29, 30, 34, 28])
after = np.array([35, 36, 38, 40, 42, 37, 39, 41, 36, 38])

# Null hypothesis: mean(after) <= mean(before)
# Perform independent two-sample t-test (one-sided test)
t_stat, p_val = stats.ttest_ind(after, before)

# Print results
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_val}")

# Conclusion
alpha = 0.05
if p_val/2 < alpha and t_stat > 0:  # One-tailed test
    print("Reject the null hypothesis: Delivery times have significantly increased.")
else:
    print("Fail to reject the null hypothesis: No significant increase in delivery times.")
```

```python
import numpy as np
import scipy.stats as stats

# Example Data: Sample scores of 30 students
sample_scores = np.array([75, 82, 79, 80, 85, 90, 88, 76, 81, 84, 83, 79, 78, 86, 77, 91, 87, 85, 78, 82, 80, 84, 89, 90, 85, 87, 92, 77, 88, 90])
population_mean = 80  # National average
population_std = 10  # Given population standard deviation
sample_size = len(sample_scores)

# Step 1: Calculate the sample mean
sample_mean = np.mean(sample_scores)

# Step 2: Calculate the Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 3: Determine the critical Z-value for a two-tailed test at alpha = 0.05
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)

# Step 4: Rejection region approach and conclusion
if abs(z_score) > critical_value:
    print(f"Reject the null hypothesis (Z-score: {z_score}, Critical value: {critical_value})")
else:
    print(f"Fail to reject the null hypothesis (Z-score: {z_score}, Critical value: {critical_value})")

print(f"Z-score: {z_score}, Critical value: ±{critical_value}")

```

### Type 1 & Type 2 Errors

**Case Study 1: Medical Testing**

**Scenario**:
A medical company has developed a new diagnostic test for a rare disease. The test is highly sensitive, meaning it rarely misses true cases of the disease (low Type II error rate). However, the company is concerned about the Type I error rate, as false positives can lead to unnecessary stress and further testing for patients.

**Assignment:**

1. Describe the potential consequences of a Type I error in this medical testing scenario.
2. Explain the potential consequences of a Type II error.
3. Discuss the trade-off between Type I and Type II errors in medical testing and how it influences the decision on test thresholds.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Simulate some medical test data (hypothetical example)
np.random.seed(0)
n_samples = 1000
# 95% of patients don't have the disease, 5% have the disease
y_true = np.random.binomial(1, 0.05, n_samples)

# Simulated test scores (logistic regression output probabilities)
X = np.random.randn(n_samples, 1)
# Model to predict probabilities
model = LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y_true, test_size=0.3)
model.fit(X_train, y_train)
y_prob = model.predict_proba(X_test)[:, 1]

# ROC Curve and AUC Score
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test, y_prob)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate (Type I Error)')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve: Trade-off Between Type I and Type II Errors')
plt.legend(loc='best')
plt.grid(True)
plt.show()

# Simulate confusion matrix at different thresholds
def print_confusion_matrix_at_threshold(threshold):
    y_pred = (y_prob >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    print(f"Threshold: {threshold}")
    print(f"Confusion Matrix: \n[[TN: {tn}, FP: {fp}], [FN: {fn}, TP: {tp}]]")
    print(f"Type I Error (False Positive Rate): {fp / (fp + tn):.2f}")
    print(f"Type II Error (False Negative Rate): {fn / (fn + tp):.2f}")
    print('-' * 50)

# Example thresholds to explore Type I and Type II errors
thresholds_to_test = [0.3, 0.5, 0.7]
for threshold in thresholds_to_test:
    print_confusion_matrix_at_threshold(threshold)

```

**Case Study 2: Criminal Justice System**

**Scenario:**
In a criminal trial, the prosecution aims to convict a defendant based on evidence. The consequences of a wrongful conviction (Type I error) are severe, but so are the consequences of letting a guilty person go free (Type II error).

**Assignment:**

1. Describe a situation in the criminal justice system where a Type I error can occur and its implications.
2. Provide an example of a situation where a Type II error can occur and its potential consequences.
3. Discuss the challenges and ethical considerations in finding a balance between minimizing Type I and Type II errors in criminal trials.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, confusion_matrix

# Simulate data for innocent and guilty individuals
np.random.seed(0)

# Suppose we have 1000 cases: 950 are innocent, 50 are guilty
n_innocent = 950
n_guilty = 50

# Model prediction probabilities for the cases
# Innocent cases have low probability of being guilty, guilty cases have high probability
innocent_scores = np.random.normal(0.3, 0.1, n_innocent)  # Innocents usually score low
guilty_scores = np.random.normal(0.7, 0.1, n_guilty)  # Guilty people usually score high

# Combine the scores
y_true = np.array([0] * n_innocent + [1] * n_guilty)  # True labels (0 = innocent, 1 = guilty)
y_scores = np.concatenate([innocent_scores, guilty_scores])  # Model scores

# Plot the distribution of scores
plt.figure(figsize=(8, 6))
plt.hist(innocent_scores, bins=30, alpha=0.5, label='Innocent (True Negative)', color='blue')
plt.hist(guilty_scores, bins=30, alpha=0.5, label='Guilty (True Positive)', color='red')
plt.axvline(x=0.5, color='black', linestyle='--', label='Decision Threshold')
plt.xlabel('Score (Probability of Guilt)')
plt.ylabel('Frequency')
plt.title('Distribution of Evidence Scores for Innocent and Guilty Individuals')
plt.legend()
plt.show()

# Define a function to print the confusion matrix and calculate Type I and Type II errors
def print_confusion_matrix_and_errors(threshold):
    y_pred = (y_scores >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    print(f"Threshold: {threshold}")
    print(f"Confusion Matrix: \n[[TN: {tn}, FP: {fp}], [FN: {fn}, TP: {tp}]]")
    print(f"Type I Error (False Positive Rate): {fp / (fp + tn):.2f}")
    print(f"Type II Error (False Negative Rate): {fn / (fn + tp):.2f}")
    print('-' * 50)

# Check confusion matrix and error rates at different decision thresholds
thresholds = [0.3, 0.5, 0.7]
for threshold in thresholds:
    print_confusion_matrix_and_errors(threshold)

# Plot ROC curve to visualize trade-off between Type I and Type II errors
fpr, tpr, _ = roc_curve(y_true, y_scores)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label='ROC Curve')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate (Type I Error)')
plt.ylabel('True Positive Rate (1 - Type II Error)')
plt.title('ROC Curve: Trade-off Between Type I and Type II Errors')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

**Case Study 2: Baseball Player Performance**

**Scenario**:
You are a data analyst for a professional baseball team. The team's management wants to assess whether a new training program has led to a significant improvement in the batting averages of their players. You have collected batting average data for two groups of players: one group who underwent the new training program (Group A) and another group who did not (Group B).

**Data**:

**Group A** (New Training Program):

Sample Size: 25 players,
Mean Batting Average: 0.285,
Sample Standard Deviation: 0.040.

**Group B** (No Training Program):

Sample Size: 30 players,
Mean Batting Average: 0.270,
Sample Standard Deviation: 0.045.

**Assignment:**

1. Formulate the null and alternate hypotheses to test whether the new training program has led to a significant improvement in batting averages.
2. Conduct a two-sample t-test using the provided data. Calculate the t-statistic.
3. Calculate the degrees of freedom for the test.
4. Calculate the p-value associated with the test.
5. Interpret the results and make a recommendation to the team's management based on the p-value approach.

```python
import numpy as np
from scipy import stats
```

```python
# Group A (New Training Program)
n_A = 25
mean_A = 0.285
std_A = 0.040

# Group B (No Training Program)
n_B = 30
mean_B = 0.270
std_B = 0.045
```

```python
# Calculate the variance
var_A = std_A**2
var_B = std_B**2

# Calculate the t-statistic
t_statistic = (mean_A - mean_B) / np.sqrt((var_A / n_A) + (var_B / n_B))
```

```python
# Calculate degrees of freedom
numerator = ((var_A / n_A) + (var_B / n_B))**2
denominator = ((var_A / n_A)**2 / (n_A - 1)) + ((var_B / n_B)**2 / (n_B - 1))
df = numerator / denominator

# Calculate the p-value
p_value = stats.t.sf(np.abs(t_statistic), df) * 2  # Two-tailed test

# Set significance level
alpha = 0.05

# Print results
print(f"T-statistic: {t_statistic}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value}")

# Make a decision
if p_value < alpha:
    print("Reject the null hypothesis: The new training program has led to a significant improvement in batting averages.")
else:
    print("Fail to reject the null hypothesis: There is no significant improvement in batting averages due to the new training program.")
```
