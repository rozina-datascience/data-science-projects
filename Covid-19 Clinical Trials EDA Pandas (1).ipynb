```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
def visualize_data(df , caption = 'Percentage of missing data in each feature'):
  missing_data= df.isnull().mean()*100
  plt.figure(figsize=(15,8))
  sns.set(rc={'figure.figsize':(15,8.27)})
  plt.xticks(rotation=90)
  fig=sns.barplot(x = missing_data.index,y=missing_data.values)
  plt.title('Percentage of missing data in each feature')
```

```python
plt.show()
visualize_data(df,'Percentage of missing data in each feature')
```
